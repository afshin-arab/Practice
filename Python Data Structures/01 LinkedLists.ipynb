```python
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
        
        
class LinkedList:
    def __init__(self):
        self.head = None
    
    def print(self):
        if self.head == None: #empty list!
            print("Linked list is empty")
            return
        
        itr = self.head 
        linked_list_str = ''
        
        while itr:
            linked_list_str += str(itr.data) + ' >> ' if itr.next else str(itr.data)
            itr = itr.next
        print(linked_list_str)    
    
    def length(self):
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
        return count
    
    def insert(self,data):
        node = Node(data,self.head)
        self.head = node  #thus self.head is a Node class
    
    def insert_end(self,data):
        if self.head == None:
            self.head = Node(data,None)
        
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node(data, None)
        
                      
    def remove_at(self,index):
        if index <0 or index >= self.length():
            raise Exception("Invalid Index: Index Out of Range")
        if index == 0:
            self.head = self.head.next
            return
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
            itr = itr.next
            count += 1
            
    def insert_at(self,index,data):
        if index <0 or index > self.length():
            raise Exception("Invalid Index: Index Out of Range")
        if index == 0:
            self.insert(data)
            return
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                node = Node(data,itr.next)
                itr.next = node
                break
            itr = itr.next
            count += 1
    
    def insert_values(self,list,intact=True): 
        self.head = None
        if intact:
            for data in reversed(list):
                self.insert(data)
        else:
            for data in list:
                self.insert(data)
        
    
    
        
```
